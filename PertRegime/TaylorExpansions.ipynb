```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def sin_exp(x,n): #returns the nth term of the taylor expansion of sin around 0
    return (-1)**n * x**(2*n+1) / np.math.factorial(2*n + 1)

def cos_exp(x,n): #returns the nth term of the taylor expansion of cos around 0
    return (-1)**n * x**(2*n) / np.math.factorial(2*n)
```

```python
n = 0
x = np.linspace(-2*np.pi,2*np.pi, 150)
y_sin = sin_exp(x,n) # first term approximation of sin
y_cos = cos_exp(x,n) # first term approximation of cos
```

```python
plt.plot(x,y_sin,"b.")
plt.plot(x,y_cos,"r.")
plt.ylim((-1,1))
n = n + 1 #consider next term
y_sin = y_sin + sin_exp(x,n) 
y_cos = y_cos + cos_exp(x,n)
```

```python
def ln_exp(x,n): #returns the nth term of the taylor expansion of ln around 1
    return (-1)**(n-1) * (x-1)**(n) / n
```

```python
n = 1
x = np.linspace(0,1, 150)
y_ln = ln_exp(x,n)
```

```python
plt.plot(x,y_ln,"k.")
n = n + 1 #consider next term
y_ln = y_ln + ln_exp(x,n) 
```

## Concavity of entropy

```python
def entropy(p_1,p_2):
    return -np.sum([ p*np.log(p) for p in [p_1, 1-p_1, p_2, 1-p_2] ])
```

```python
N = 20
p_1 = np.outer(np.linspace(0.1, 0.9, N), np.ones(N))
p_2 = p_1.copy().T
H = np.zeros_like(p_1)
print(p_1.shape, H.shape)
for i in range(N):
    for j in range(N):
        H[i,j] = entropy(p_1[i,j],p_2[i,j])
```

```python
fig = plt.figure()
 
# syntax for 3-D plotting
ax = plt.axes(projection ='3d')
 
# syntax for plotting
ax.plot_surface(p_1, p_2, H, cmap ='viridis', edgecolor ='green')
ax.set_title('Entropy over 2 variables')
ax.set_xlabel('P_1')
ax.set_ylabel('P_1')
plt.show()
```
